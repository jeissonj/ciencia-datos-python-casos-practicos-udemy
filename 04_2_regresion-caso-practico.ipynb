```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
datos_dummies.columns
```

```python
lm = LinearRegression()
X = datos_dummies[['Potencia de Salida (W)', 'Tamaño casa (m2',
       'Panel Solar_Con Dos Ejes de Movimiento',
       'Panel Solar_Con Eje de Seguimiento', 'Panel Solar_Estandar',
       'Tamano(m)_1.7x1', 'Tamano(m)_4x2', 'Tamano(m)_8x4',
       'Ubicació_Con obstruciones', 'Ubicació_Sin obstruciones ',
       'Ubicación_clasifiacio_Buena ', 'Ubicación_clasifiacio_Mala',
       'Inclinacion_Hacia el sur', 'Inclinacion_Latitud pesima ',
       'Inclinacion_clasifiacion_Buena', 'Inclinacion_clasifiacion_Buena ',
       'Inclinacion_clasifiacion_Buenaa', 'Inclinacion_clasifiacion_Mala',
       'Panel Solar2_Aprobado', 'Panel Solar2_Denegado ']]
y = datos_dummies['Precio (MXN)']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
lm.fit(X_train, y_train)
prediccion = lm.predict(X_test)
```

```python
resultado = {'Real':y_test,"Prediccion":prediccion}
resultado = pd.DataFrame(resultado)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
Mae = mean_absolute_error(y_test,prediccion)
R_ = lm.score(X_train,y_train)
print(f"El mae es de {Mae} y el valor de R2 es {R_}")
```
