```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ruta = "paneles_solares_costos_por_tecnologia.csv"
df = pd.read_csv(ruta)
```

```python
plt.scatter(df["Precio (MXN)"], df["Potencia de Salida (W)"])
plt.xlabel("Precio")
plt.ylabel("Potencia de Salida")
plt.title("Paneles solares")
plt.show()
```

```python
# df.describe()
# df.info()
sns.barplot(x="Panel Solar", y="Precio (MXN)", data=df, hue="Panel Solar2")
```
