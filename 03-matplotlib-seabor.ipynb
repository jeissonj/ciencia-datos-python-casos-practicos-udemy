```python
tips = sns.load_dataset("tips")
average_tip_by_day = tips.groupby("day")["tip"].mean().reset_index()
sns.barplot(x = "day", y ="tip", data = average_tip_by_day)
plt.title("promedio de propias por dia de la semana")
plt.xlabel("D√≠a de la semana")
plt.ylabel("Promedio de propina")
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(x = "day", y = "tip", data=tips)
plt.title("Distribuci√≥n")
plt.xlabel("D√≠a de la seman")
plt.ylabel("Propina")
plt.show()
```

```python
# Curvas de nivel
def f(x,y):
    return np.sin(x) ** 2 + np.cos(5 + x*y) + 2*np.cos(x)
x = np.linspace(0,5,100)
y = np.linspace(0,5,100)
x,y = np.meshgrid(x,y)
z = f(x,y)
```

```python
plt.contour(x,y,z)
plt.show()
```

```python
plt.contour(x,y,z, levels = 15)
plt.show()
```
