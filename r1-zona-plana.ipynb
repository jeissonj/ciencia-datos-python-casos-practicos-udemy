```python
import matplotlib.pylab as plt
import numpy as np
import math as mt
import seaborn as sn
import sklearn 
```

```python
x = np.linspace(1,20,20)
y = np.array(list(map(lambda i: mt.exp(i),x)))
```

```python
fig = plt.figure()
plt.plot(x,y,label="Original")
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Titulo principal")
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Supongamos que tienes tus datos experimentales en dos listas o arrays de NumPy: x y y

# Identificar la zona plana (por ejemplo, asumiendo que los datos están en la primera parte del array)
indice_inicio = 0
indice_final = 50

# Seleccionar los puntos correspondientes a la zona plana
x_plano = x[indice_inicio:indice_final]
y_plano = y[indice_inicio:indice_final]

# Realizar la regresión lineal
modelo = LinearRegression()
modelo.fit(x_plano.reshape(-1, 1), y_plano)  # Ajustar el modelo a los datos

# Obtener los coeficientes de la regresión
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Graficar los datos experimentales y la regresión lineal
plt.scatter(x, y, label='Datos experimentales')
plt.plot(x_plano, modelo.predict(x_plano.reshape(-1, 1)), color='red', label='Regresión lineal en zona plana')
plt.xlabel('Variable independiente')
plt.ylabel('Variable dependiente')
plt.title('Regresión lineal en zona plana de datos experimentales')
plt.legend()
plt.show()

# Imprimir los coeficientes de la regresión
print("Pendiente:", pendiente)
print("Intercepto:", intercepto)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Supongamos que tus datos experimentales están en dos arrays de NumPy: x y y

# Calcular la pendiente entre puntos adyacentes
pendientes = np.diff(y) / np.diff(x)

# Definir un umbral para la pendiente
umbral_pendiente = 2.06641309e+06  # Ajusta este valor según tus datos y criterios
 
# Encontrar los índices de los puntos donde la pendiente es menor que el umbral
indices_zona_plana = np.where(pendientes < umbral_pendiente)[0]

# Seleccionar los puntos correspondientes a la zona plana
x_zona_plana = x[indices_zona_plana]
y_zona_plana = y[indices_zona_plana]


# Realizar la regresión lineal
modelo = LinearRegression()
modelo.fit(x_zona_plana.reshape(-1, 1), y_zona_plana)  # Ajustar el modelo a los datos
# Obtener los coeficientes de la regresión
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_


# Visualizar los datos experimentales y la zona plana identificada
import matplotlib.pyplot as plt

plt.plot(x, y, label='Datos experimentales')
plt.plot(x_plano, modelo.predict(x_plano.reshape(-1, 1)), color='red', label='Regresión lineal en zona plana')
plt.scatter(x_zona_plana, y_zona_plana, color='red', label='Zona plana identificada')

plt.xlabel('Variable independiente')
plt.ylabel('Variable dependiente')
plt.title('Identificación de la zona plana en datos experimentales')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Supongamos que tienes tus datos experimentales en dos listas o arrays de NumPy: x y y

# Definir una función lineal para el ajuste
def funcion_lineal(x, m, b):
    return m * x + b

# Ajustar la curva a los datos experimentales
parametros, covarianzas = curve_fit(funcion_lineal, x, y)

# Calcular los valores ajustados usando los parámetros obtenidos
valores_ajustados = funcion_lineal(x, *parametros)

# Calcular el error residual para cada punto
errores_residuales = np.abs(y - valores_ajustados)

# Definir un umbral para el error residual
umbral_error = 0.1  # Ajusta este valor según tus datos y criterios

# Encontrar los índices de los puntos donde el error residual es menor que el umbral
indices_zona_plana = np.where(errores_residuales < umbral_error)[0]

# Seleccionar los puntos correspondientes a la zona plana
x_zona_plana = x[indices_zona_plana]
y_zona_plana = y[indices_zona_plana]

# Visualizar los datos experimentales y la zona plana identificada
plt.scatter(x, y, label='Datos experimentales')
plt.plot(x, valores_ajustados, color='red', label='Ajuste lineal')
plt.scatter(x_zona_plana, y_zona_plana, color='green', label='Zona plana identificada')
plt.xlabel('Variable independiente')
plt.ylabel('Variable dependiente')
plt.title('Identificación de la zona plana en datos experimentales')
plt.legend()
plt.show()
```
